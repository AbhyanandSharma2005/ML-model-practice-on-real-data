```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import StandardScaler
import seaborn as sns
import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data
y = iris.target

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42, stratify=y)

model = RandomForestClassifier(
    n_estimators=100,
    random_state=42,
    max_features='sqrt'
)
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=iris.target_names, yticklabels=iris.target_names)
plt.title('Random Forest (Bagging) Confusion Matrix (Iris Dataset)')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()

print(f"Random Forest Accuracy on training set: {model.score(X_train, y_train):.4f}")
print(f"Random Forest Accuracy on test set: {model.score(X_test, y_test):.4f}")
```
